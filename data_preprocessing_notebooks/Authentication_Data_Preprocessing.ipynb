# Authentication Feature Engineering

```python
# Failed login indicator
auth_df['is_failed_login'] = (~auth_df['success']).astype(int)

# MFA disabled indicator
auth_df['mfa_disabled'] = (~auth_df['mfa_enabled']).astype(int)

# Token expired indicator
auth_df['token_expired_flag'] = auth_df['token_expired'].astype(int)

# Blocked indicator
auth_df['blocked_flag'] = auth_df['blocked'].astype(int)

# Suspicious activity indicator
auth_df['suspicious_flag'] = auth_df['suspicious_activity'].astype(int)
```

```python
# Weekend login
auth_df['is_weekend'] = auth_df['day_of_week'].isin([5, 6]).astype(int)

# Late-night login (midnight to 5 AM)
auth_df['late_night_login'] = (
    (auth_df['hour'] >= 0) &
    (auth_df['hour'] <= 5)
).astype(int)
```

```python
auth_features = auth_df[
[
'failed_attempts',
'session_duration',
'password_age_days',
'privilege_level',
'threat_level',

'hour',
'day_of_week',
'month',

'is_failed_login',
'mfa_disabled',
'token_expired_flag',
'blocked_flag',
'suspicious_flag',
'is_weekend',
'late_night_login'
]
]
```

```python
from sklearn.preprocessing import StandardScaler
import joblib
```

```python
scaler = StandardScaler()
```

```python
auth_scaled = scaler.fit_transform(auth_features)
```
