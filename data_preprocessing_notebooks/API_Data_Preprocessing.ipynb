```python
print("Request duplicates:", request_df.duplicated().sum())
print("Response duplicates:", response_df.duplicated().sum())
print("Error duplicates:", error_df.duplicated().sum())
```

```python
request_df = request_df.drop_duplicates()
print(request_df.shape)
```

```python
request_df.isnull().sum()
```

```python
request_df['errorCode'] = request_df['errorCode'].fillna('NoError')
```

```python
response_df[['message',
             'file',
             'line',
             'logger_name',
             'level']].isnull().sum()
```

```python
response_df = response_df.drop(
    columns=[
        'message',
        'file',
        'line',
        'logger_name',
        'level'
    ]
)
```

```python
response_df['method'] = response_df['method'].fillna('Unknown')

response_df['path'] = response_df['path'].fillna('Unknown')

response_df['status_code'] = response_df['status_code'].fillna(0)

response_df['duration_ms'] = response_df['duration_ms'].fillna(
    response_df['duration_ms'].median()
)
```

```python
error_df.isnull().sum()
```

```python
error_df['error_type'] = error_df['error_type'].fillna('NoError')
```

```python
print(request_df.isnull().sum().sort_values(ascending=False).head())

print(response_df.isnull().sum().sort_values(ascending=False).head())

print(error_df.isnull().sum().sort_values(ascending=False).head())
```

```python
request_df['eventTime'] = pd.to_datetime(
    request_df['eventTime']
)

response_df['timestamp'] = pd.to_datetime(
    response_df['timestamp']
)

error_df['timestamp'] = pd.to_datetime(
    error_df['timestamp']
)
```

```python
request_df['hour'] = request_df['eventTime'].dt.hour
request_df['dayofweek'] = request_df['eventTime'].dt.dayofweek

response_df['hour'] = response_df['timestamp'].dt.hour
response_df['dayofweek'] = response_df['timestamp'].dt.dayofweek

error_df['hour'] = error_df['timestamp'].dt.hour
error_df['dayofweek'] = error_df['timestamp'].dt.dayofweek
```

```python
request_df['is_error'] = (
    request_df['errorCode'] != 'NoError'
).astype(int)

error_df['is_error'] = (
    error_df['status_code'] >= 400
).astype(int)
```

```python
request_features = (
    request_df.groupby('sourceIPAddress')
    .agg(
        request_count=('eventID','count'),
        unique_events=('eventName','nunique'),
        unique_services=('eventSource','nunique'),
        error_count=('is_error','sum'),
        active_hours=('hour','nunique')
    )
    .reset_index()
)

request_features.head()
```

```python
response_features = (
    response_df.groupby('instance_id')
    .agg(
        avg_duration=('duration_ms','mean'),
        max_duration=('duration_ms','max'),
        total_events=('event_count','sum'),
        unique_paths=('path','nunique')
    )
    .reset_index()
)

response_features.head()
```

```python
error_features = (
    error_df.groupby('client_ip')
    .agg(
        error_count=('is_error','sum'),
        avg_latency=('latency_ms','mean'),
        retry_sum=('retry_count','sum'),
        unique_endpoints=('endpoint','nunique')
    )
    .reset_index()
)

error_features.head()
```

```python
print(request_features.shape)
print(response_features.shape)
print(error_features.shape)
```

```python
request_features.describe()

response_features.describe()

error_features.describe()
```

```python
print("Unique instance IDs:", response_df['instance_id'].nunique())

response_df['instance_id'].value_counts().head()
```

```python
request_features_extra = (
    request_df.groupby('sourceIPAddress')
    .agg(
        unique_users=('userIdentityuserName', 'nunique'),
        unique_regions=('awsRegion', 'nunique'),
        unique_agents=('userAgent', 'nunique')
    )
    .reset_index()
)
```

```python
request_features = request_features.merge(
    request_features_extra,
    on='sourceIPAddress',
    how='left'
)
```

```python
print(request_features.shape)
```

```python
request_features.head()
```

```python
request_features['error_rate'] = (
    request_features['error_count']
    /
    request_features['request_count']
)
```

```python
request_features[['error_count',
                  'request_count',
                  'error_rate']].head()
```

```python
request_features.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
api_features = request_features.copy()
```

```python
feature_cols = [
    'request_count',
    'unique_events',
    'unique_services',
    'error_count',
    'active_hours',
    'unique_users',
    'unique_regions',
    'unique_agents',
    'error_rate'
]
```

```python
scaler = StandardScaler()

api_features[feature_cols] = scaler.fit_transform(
    api_features[feature_cols]
)
```

```python
api_features = api_features.drop(
    columns=['active_hours']
)
```
